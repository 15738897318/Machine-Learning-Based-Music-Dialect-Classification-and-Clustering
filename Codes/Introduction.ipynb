```python
import numpy, scipy, matplotlib.pyplot as plt, pandas, librosa,sklearn,IPython.display as ipd
import matplotlib.style as ms
import librosa
import librosa.display
```

```python
sampling_rate=22050

x_bandari,sr=librosa.load("bandari.wav",sr=sampling_rate,mono=True,duration=30)
x_gilaki,sr=librosa.load("gilaki.wav",sr=sampling_rate,mono=True,duration=30)
x_kurdish,sr=librosa.load("kurdish.wav",sr=sampling_rate,mono=True,duration=30)
x_lori,sr=librosa.load("lori.wav",sr=sampling_rate,mono=True,duration=30)
x_turkish,sr=librosa.load("turkish.wav",sr=sampling_rate,mono=True,duration=30)
```

## Time Plot

```python
fig=plt.figure(figsize=(20,20))
fig.subplots_adjust(hspace=.5)

plt.subplot(5, 1, 1)
plt.xlabel("Time")
plt.ylabel("Amplitude")
librosa.display.waveplot(x_bandari,sampling_rate)
plt.title('Bandari')

plt.subplot(5, 1, 2)
# plt.ylim(-0.4,0.4)
plt.xlabel("Time")
plt.ylabel("Amplitude")
librosa.display.waveplot(x_gilaki,sampling_rate)
plt.title('Gilaki')

plt.subplot(5, 1, 3)
plt.xlabel("Time")
plt.ylabel("Amplitude")
librosa.display.waveplot(x_kurdish,sampling_rate)
plt.title('kurdish')

plt.subplot(5, 1, 4)
plt.xlabel("Time")
plt.ylabel("Amplitude")
librosa.display.waveplot(x_lori,sampling_rate)
plt.title('lori')

plt.subplot(5, 1, 5)
plt.xlabel("Time")
plt.ylabel("Amplitude")
librosa.display.waveplot(x_turkish,sampling_rate)
plt.title('turkish')

plt.savefig('time_plot.png')
```

## Onset Detection

```python
hop_size=512
frame_size=2048

#Onset Detection
onset_frames = librosa.onset.onset_detect(x_bandari, sr=sampling_rate,hop_length=hop_size)
onset_times = librosa.frames_to_time(onset_frames)
x=onset_frames.size
print("Number of detected onsets for Bandari song={}".format(x))
print(onset_frames)

onset_frames2 = librosa.onset.onset_detect(x_gilaki, sr=sampling_rate,hop_length=hop_size)
onset_times2 = librosa.frames_to_time(onset_frames2)
x2=onset_frames2.size
print("\nNumber of detected onsets for Gilaki song={}".format(x2))
print(onset_frames2)

onset_frames3 = librosa.onset.onset_detect(x_kurdish, sr=sampling_rate,hop_length=hop_size)
onset_times3 = librosa.frames_to_time(onset_frames3)
x3=onset_frames3.size
print("\nNumber of detected onsets for Kurdish song={}".format(x3))
print(onset_frames3)

onset_frames4 = librosa.onset.onset_detect(x_lori, sr=sampling_rate,hop_length=hop_size)
onset_times4 = librosa.frames_to_time(onset_frames4)
x4=onset_frames4.size
print("\nNumber of detected onsets for Lori song={}".format(x4))
print(onset_frames4)

onset_frames5 = librosa.onset.onset_detect(x_turkish, sr=sampling_rate,hop_length=hop_size)
onset_times5 = librosa.frames_to_time(onset_frames5)
x5=onset_frames5.size
print("\nNumber of detected onsets for Turkish song={}".format(x5))
print(onset_frames5)
```

##  FFT

```python
# Fourier Transform and Spectrum

fig=plt.figure(figsize=(20,20))
fig.subplots_adjust(hspace=.5)

plt.subplot(5,1,1)
FFT_bandari=scipy.fft.fft(x_bandari)
FFT_bandari_Mag=numpy.absolute(FFT_bandari)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title('Bandari')
plt.plot(FFT_bandari_Mag)

plt.subplot(5,1,2)
FFT_gilaki=scipy.fft.fft(x_gilaki)
FFT_gilaki_Mag=numpy.absolute(FFT_gilaki)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title('Gilaki')
plt.plot(FFT_gilaki_Mag)

plt.subplot(5,1,3)
FFT_kurdish=scipy.fft.fft(x_kurdish)
FFT_kurdish_Mag=numpy.absolute(FFT_kurdish)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title('Kurdish')
plt.plot(FFT_kurdish_Mag)

plt.subplot(5,1,4)
FFT_lori=scipy.fft.fft(x_lori)
FFT_lori_Mag=numpy.absolute(FFT_lori)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title('Lori')
plt.plot(FFT_lori_Mag)

plt.subplot(5,1,5)
FFT_turkish=scipy.fft.fft(x_turkish)
FFT_turkish_Mag=numpy.absolute(FFT_turkish)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title('Turkish')
plt.plot(FFT_turkish_Mag)

plt.savefig('fft.png')
```

## Spectogram

```python
#Short-Time Fourier Transform and Spectogram

window_size=2048

fig=plt.figure(figsize=(20,20))
fig.subplots_adjust(hspace=.5)

plt.subplot(5,1,1)
S = librosa.feature.melspectrogram(x_bandari, sr=sampling_rate, n_fft=window_size)
logS = librosa.amplitude_to_db(S)
plt.title('Bandari')
librosa.display.specshow(logS, sr=sampling_rate, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')

plt.subplot(5,1,2)
S2 = librosa.feature.melspectrogram(x_gilaki, sr=sampling_rate, n_fft=window_size)
logS2 = librosa.amplitude_to_db(S2)
plt.title('Gilaki')
librosa.display.specshow(logS2, sr=sampling_rate, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')

plt.subplot(5,1,3)
S3 = librosa.feature.melspectrogram(x_kurdish, sr=sampling_rate, n_fft=window_size)
logS3 = librosa.amplitude_to_db(S3)
plt.title('Kurdish')
librosa.display.specshow(logS3, sr=sampling_rate, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')

plt.subplot(5,1,4)
S4 = librosa.feature.melspectrogram(x_lori, sr=sampling_rate, n_fft=window_size)
logS4 = librosa.amplitude_to_db(S4)
plt.title('Lori')
librosa.display.specshow(logS4, sr=sampling_rate, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')

plt.subplot(5,1,5)
S5 = librosa.feature.melspectrogram(x_turkish, sr=sampling_rate, n_fft=window_size)
logS5 = librosa.amplitude_to_db(S5)
plt.title('Turkish')
librosa.display.specshow(logS5, sr=sampling_rate, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')

plt.savefig('spectogram.png')
```
